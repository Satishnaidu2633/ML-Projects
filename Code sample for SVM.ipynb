```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
#Load Data
iris= datasets.load_iris()
X = iris.data[:, :2] #Taking first two features for visualization
y = iris.target
```

```python
#Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state=42)
```

```python
#Create and train the SVM model
svm_model = SVC(kernel = 'rbf', C=1.0, gamma='scale') #gamma=1/alpha
svm_model.fit(X_train, y_train)
```

```python
#Make predictions
y_pred = svm_model.predict(X_test)
```

```python
#Evaluate Model
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

```python
#Plot decision Boundary
def plot_decision_boundary(X, y, model):
    h = 0.02 #step size in the mess grid
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    plt.contourf(xx, yy, Z, aplha = 0.3)
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k', marker='o', cmap=plt.cm.Set1)
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.title("SVM Decision Boundary")
    plt.show()
plot_decision_boundary(X, y, svm_model)
```
